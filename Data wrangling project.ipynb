# Remove all duplicates observation

```python
df.duplicated()
```

```python
#creating a backup
df_1=df.copy()

#removing duplicates
df_1=df_1.drop_duplicates()
```

# Find the missing values in all columns

```python
df_1.isna()
```

```python
df_1.isna().sum()
```

# Remove all observatioms where continent columns value is missing

```python
df_1['continent'].isna()
```

```python
# Using subset parameter in dropna
```

```python
df_1=df_1.dropna(subset=['continent'])
```

```python
df_1['continent'].isna()
```

```python
df_1['continent'].isna().sum()
```

# Fill all missing values with 0


```python
df_1=df_1.fillna(0)
df_1
```

# Convert date column in datetime format using pandas.to _datetime

```python
df_1['date']=pd.to_datetime(df_1['date'])
```

```python
df_1['date']
```

# Create new column month after extracting month data from date column


```python
df_1['month']=pd.DatetimeIndex(df_1['date']).month
```

```python
df_1['month']
```

# Find max value in all columns using groupby function on 'continent' colummn

```python
df_1.groupby('continent').max().reset_index()
```

```python
df_1
```

# Store the result in a new datafrome named 'df_groupby'.(Use df_groupby dataframe for all further analysis)

```python
df_groupby=df_1
```

```python
df_groupby
```

# Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases'

```python
df_groupby['total_deaths_to_total_cases']=df_groupby["total_deaths"]/df_groupby["total_cases"]
df_groupby
```
